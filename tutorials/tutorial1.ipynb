```python
import matplotlib.pyplot as plt
import numpy as np
import soundfile as soundf
from wav2aud_zpk import wav2aud
from pathlib import Path
```

```python
# Get test_sounds directory
test_sounds_dir = Path.cwd().parent / 'test_sounds'
filters = Path.cwd().parent / 'cochba_filters.npz'
    
# Get list of sound files (excluding subdirectories)
sound_files = sorted([f.name for f in test_sounds_dir.glob('*') if f.is_file()])
print(sound_files)

# MANUALLY CHANGE
sound_file_path = test_sounds_dir / 'speech_cut.wav'

    
# params taken from matlab
sf = 16000 # target sample rate
paras = [4, 0, -2, np.log2(sf/16000), 0, 0, 1]

# test with audio files
soundData, sample_rate = soundf.read(sound_file_path)
spect = wav2aud(soundData, paras, COCHBA_filename=filters)
plt.imshow(np.flipud(spect.T), aspect='auto', origin='lower')

plt.colorbar()
plt.plot(spect)
plt.show()
```
