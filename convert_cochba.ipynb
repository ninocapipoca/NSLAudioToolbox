```python
# import cochba
df = pd.read_csv("cochba_py.csv", header=None)
df.head()
```

```python
def convert_complex(complex_str):
    complex_str = complex_str.replace('i', 'j')
    complex_str = complex_str.replace(' ', '')
    complex_str = complex_str.replace('+-', '-')
    return complex_str
```

```python
df_processed = df.applymap(convert_complex)
df_processed
```

```python
# export as csv 
df_processed.to_csv("cochba_py_processed.csv", index=False, header=False)
```
